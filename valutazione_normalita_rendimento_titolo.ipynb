# VALUTARE SE LA DISTRIBUZIONE DEL RENDIMENTO DI UN TITOLO SIA DI TIPO NORMALE



```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

start_date = "2020-01-01" #Inserire la data di inizio
end_date = "2022-01-01" #Inserire la data di fine
stock_symbol = "ENI.MI"  # Inserire il simbolo di Borsa per il titolo da ricercare su Yahoo Finance

data = yf.download(stock_symbol, start=start_date, end=end_date)["Adj Close"]

daily_returns = data.pct_change().dropna()

mean_return = daily_returns.mean()
std_return = daily_returns.std()

x = np.linspace(min(daily_returns), max(daily_returns), 100)
normal_distribution = norm.pdf(x, mean_return, std_return)

plt.figure(figsize=(10, 6))
plt.hist(daily_returns, bins=40, density=True, alpha=0.6, color='blue', label='Rendimenti giornalieri')
plt.plot(x, normal_distribution, color='red', label='Distribuzione Normale')
plt.xlabel('Rendimento')
plt.ylabel('Densità')
plt.title('Distribuzione dei rendimenti del titolo ENI (01/01/2020-01/01/2022)')
plt.legend()
plt.show()
```
