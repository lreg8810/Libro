# SIMULAZIONE DEL MOTO GEOMETRICO BROWNIANO PER IL TITOLO ENI



```python
import numpy as np
import matplotlib.pyplot as plt

mean_return = 0.05 # Inserire il parametro mu
std_return = 0.2 # Inserire il parametro sigma
initial_price = 100  # Inserire il prezzo iniziale del titolo
num_paths = 10 # Inserire il numero di processi da simulare
num_years = 3 # Inserire il numero di anni oggetto di simulazione
num_steps = 252 * num_years  # 252 giorni lavorativi all'anno

dt = 1 / 252
increments = np.random.normal(mean_return * dt, std_return * np.sqrt(dt), size=(num_steps, num_paths))

prices = np.zeros((num_steps + 1, num_paths))
prices[0, :] = initial_price

for i in range(1, num_steps + 1):
    prices[i, :] = prices[i - 1, :] * np.exp(increments[i - 1, :])

plt.figure(figsize=(10, 6))
plt.plot(np.arange(0, num_steps + 1) / 252, prices)
plt.xlabel('Tempo(Anni)')
plt.ylabel('Prezzo')
plt.title('Simulazione del Moto Geometrico Browniano per il titolo ENI')
plt.show()
```
